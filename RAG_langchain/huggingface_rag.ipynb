```python
from langchain_community.document_loaders import UnstructuredURLLoader
```

```python
urls = ['https://www.cricbuzz.com/cricket-news/133516/fakhar-zaman-set-to-be-ruled-out-champions-trophy-2025',
        'https://www.cricbuzz.com/cricket-news/133512/tom-latham-will-young-crash-pakistan-party-in-2025-champions-trophy-opener',
        'https://www.cricbuzz.com/cricket-news/133510/indias-bowling-balance-in-spotlight-as-dubai-weather-plays-wildcard']
loader = UnstructuredURLLoader(urls=urls)
data = loader.load()
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

```

```python
# split data
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000)
docs = text_splitter.split_documents(data)
```

```python
print(len(docs))
```

```python


print(docs[0].page_content)
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings()
```

```python
embed_vector = embeddings.embed_query("what is 2+3?")
print(embed_vector)
```

```python

from langchain.vectorstores import Chroma

# vectorstore = Chroma.from_documents(documents=docs, persist_directory=r"D:\RAG_langchain\vector_hugging",embedding=embeddings)
vectorstore = Chroma.from_documents(documents=docs, embedding=embeddings)
```

```python
text_gen_pipeline = pipeline("text-generation",
    model=model_id,
    tokenizer=model_id,
    torch_dtype=torch.bfloat16,
    trust_remote_code=True,
    max_new_tokens =400,
    device='cpu')
```

```python
llm = HuggingFacePipeline(pipeline=text_gen_pipeline,batch_size=1)
```

```python
from langchain.prompts import PromptTemplate
```

```python
prompttemplate = """
<System>
Answer the question based on your knowledge.

<Question>
{question}
</Question>
"""
```

```python
prompt=PromptTemplate(template=prompttemplate, input_variables=["question"])
```

```python
from langchain_core.output_parsers import StrOutputParser
```

```python
chain = prompt|llm|StrOutputParser()
```

```python
Out = chain.invoke({"question", "What are Large Language Models?"})
```

```python
prompttemplate = """
<System>
Answer the question based on below context.

<context>
{context}
<context>

<Question>
{question}
</Question>
"""
```

```python
prompt=PromptTemplate(template=prompttemplate, input_variables=["context","question"])
```

```python
from langchain_core.runnables import RunnablePassthrough
```

```python
rag_chain = {"context": retriver, "question": RunnablePassthrough()} | prompt | llm | StrOutputParser()
```

```python
question ="Who got injured in CHampions Trophy?"
```

```python
answer = rag_chain.invoke(question)
```

```python

# Making the response readable
answer = answer.replace("</s>", "").strip()
print("Response:", answer)
```

```python
model_name = "EleutherAI/gpt-neo-1.3B"
model = AutoModelForCausalLM.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)

text_generation_pipeline = pipeline(
    model=model,
    tokenizer=tokenizer,
    task="text-generation",
    temperature=0.1,
    do_sample=True,
    repetition_penalty=1.1,
    return_full_text=True,
    max_new_tokens=400,
)
llm = HuggingFacePipeline(pipeline=text_generation_pipeline)
```
